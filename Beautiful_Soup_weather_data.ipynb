```python
import requests
from bs4 import BeautifulSoup
url = 'http://www.weather.com.cn/weather/101210410.shtml'
header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}
response = requests.get(url,headers=header)
response.encoding='utf-8' #Since this html page contains chinese characters, need to have this encoding for it to be rendered correctly
soup = BeautifulSoup(response.text,'lxml')
print(soup.prettify())
```

```python
ul_tag= soup.find('ul','t clearfix') # using a combination of node and classname
li_tag = ul_tag.find_all('li') #returning a list
li_tag[0]
```

```python
for tag in li_tag:
    print(tag.h1.text)
    print(tag.find('p','wea').text)
    print(tag.find('p','tem').span.text + '℃')
    print(tag.find('p','tem').i.text)
    print('--------------------------')
```
